```python
import pandas as pd


csv_file_path = 'betweenness_partial.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Define the limits for rows
m_limit = 111
n_limit = 222

# Create an empty DataFrame to store differences
all_differences_df = pd.DataFrame()

# Iterate through rows
for m in range(3, m_limit):
    # Calculate the index of the corresponding row in the second range
    n = m + (n_limit - m_limit)

    # Check if n is within the limits
    if n <= df.shape[0]:
        # Extract values for rows m and n
        row_m_values = pd.to_numeric(df.iloc[m - 1].dropna(), errors='coerce')
        row_n_values = pd.to_numeric(df.iloc[n - 1].dropna(), errors='coerce')

        # Calculate the differences between corresponding values
        differences = row_m_values - row_n_values

        # Create a DataFrame for the current pair of rows
        differences_df = pd.DataFrame({
            'Column': differences.index,
            f'Row{m}_Value': row_m_values.values,
            f'Row{n}_Value': row_n_values.values,
            'Difference': differences.values
        })

        # Concatenate the current DataFrame to the overall DataFrame
        all_differences_df = pd.concat([all_differences_df, differences_df], ignore_index=True)

# Display the DataFrame of differences
print("Differences between Rows 4 to 111 and Rows 113 to 225:")
print(all_differences_df)
```
